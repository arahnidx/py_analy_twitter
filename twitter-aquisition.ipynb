```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

# Creating list to append tweet data to
tweets_list2 = []

users_name = ['BTC_Archive','Chainalysis', 'CryptoQuant', 'cryptoquant_com', 'Deplhi_Digital', 'glassnodealerts', 'MessariCrypto', 'nomicsfinance', 'santimentfeed', 'skewdotcom', 'whale_alert', 'futofmoney', 'BarrySilbert', 'CathieDWood', 'CZ_Binance', 'Dan_Pantera', 'Jack', 'Saylor', 'RaoulGMI', 'twobitidiot', 'SBF_FTX', 'Cameron', 'Tyler', 'ElonMusk', 'CryptoCoinsNews', 'crypto', 'BTCNewsletter', 'coindesk', 'crypto', 'BTCTN']
for n, k in enumerate(users_name):
# Using TwitterSearchScraper to scrape data and append tweets to list'from:{}'.format(users_name[n])
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper('btc since:2021-10-01 until:2021-10-31 from:{}'.format(users_name[n])).get_items()):
        if i>150:
            break
        tweets_list2.append([tweet.date, tweet.id, tweet.content, tweet.user.username,tweet.user.followersCount, tweet.likeCount, tweet.retweetCount])

# Creating a dataframe from the tweets list above
tweets_df2 = pd.DataFrame(tweets_list2, columns=['Datetime', 'Tweet Id', 'Text', 'Username', 'Followers', 'Likes', 'Retweets'])
```
